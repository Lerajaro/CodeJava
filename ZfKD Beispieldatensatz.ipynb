```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Function to generate random birthdates following a Gaussian distribution
def random_gaussian_birthdates(mean_date, std_dev, size):
    random_days = np.random.normal(0, std_dev, size=size)
    birth_dates = [mean_date + timedelta(days=int(days)) for days in random_days]
    return birth_dates

# Number of birthdates you want to generate
num_birthdates = 1000

# Define the mean birthdate and standard deviation for the Gaussian distribution
mean_birthdate = datetime(1980, 1, 1)
standard_deviation = 1500  # Adjust this value to control the spread of the birthdates

# Generate random birthdates following a Gaussian distribution
birth_dates = random_gaussian_birthdates(mean_birthdate, standard_deviation, num_birthdates)

# Convert birthdates to numerical values (days since a reference date)
reference_date = datetime(1980, 1, 1)
numerical_birthdates = [(bd - reference_date).days for bd in birth_dates]

# Create a histogram to visualize the distribution
plt.hist(numerical_birthdates, bins=50, edgecolor='black')
plt.xlabel('Days since January 1, 1980')
plt.ylabel('Frequency')
plt.title('Birthdates Distribution')
plt.grid(True)
plt.show()

```
